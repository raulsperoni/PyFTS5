```python
import json

import fts

with open("100015_216865056237327204.json") as f:
    data = json.load(f)
```

```python
db = fts.FullTextSearchDB()
for i, page in enumerate(data["pages"]):
    db.add_document(doc_id=page["page_number"], content=page["text"])
```

```python
for i, h, o in db.search_near("expert stokely", max_distance=10, highlight=True):
    print(h)
```
